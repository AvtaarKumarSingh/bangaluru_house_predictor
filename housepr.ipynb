```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Bengaluru_House_Data.csv')
data.drop(columns=['area_type','availability','availability'],inplace=True)
```

```python
data
```

```python
data.dropna(inplace=True)
data.location.value_counts()
data['location'] = data['location'].apply(lambda x:x.strip())
```

```python
location_new=data.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_new
```

```python
location_less_than_10 =location_new[location_new <=10]
```

```python
location_less_than_10
```

```python
data['location'] = data['location'].apply(lambda x:'other' if x in location_less_than_10 else x)
```

```python
data['location'].value_counts()
```

```python
data
```

```python
data['size'].unique()
```

```python
data['bedrooms']=data['size'].apply(lambda x:x.split(' ')[0])
```

```python
data
```

```python
data.total_sqft.unique()
```

```python
def clean(sqft):
    if isinstance(sqft, (int, float)):
        return float(sqft)

    tokens = str(sqft).split('-')

    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2

    try:
        return float(sqft)
    except:
        return None
```

```python
clean('200')
```

```python
data['total_sqft']=data['total_sqft'].apply(clean)
data.total_sqft.unique()
```

```python
data.dropna(inplace=True)
data.describe()
```

```python
data['bedrooms'] = data['bedrooms'].astype(int)
```

```python
data['sqft_per_bed'] = data['total_sqft'] / data['bedrooms']
```

```python
data.sqft_per_bed.describe()
```

```python
data1=data[data['sqft_per_bed'] >=300]
```

```python
data1
```

```python
data1['price_per_sqft'] = data1['price']*100000/data1['total_sqft']
```

```python
data1
```

```python
data2 = data1[data1['price_per_sqft'] > 2000]
```

```python
data2.drop(columns = ['size','price_per_sqft','sqft_per_bed'],axis =1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer

col_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),
    remainder='passthrough'
)
```

```python
lr = LinearRegression()
scaler = StandardScaler()
model = make_pipeline(col_trans,scaler,lr)
```

```python
X = data2.drop(columns=['price'])
y = data2['price']
```

```python
X.select_dtypes(include='object').columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)

score = r2_score(y_test, y_pred)
print(score)
```

```python
model.score(X_test,y_test)
```

```python
import pickle as pk
pk.dump(model,open('House_prediction.pkl','wb'))
```
